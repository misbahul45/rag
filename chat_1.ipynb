```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.schema import Document
from IPython.display import display, Markdown
import os

# === Setup ===
load_dotenv()
api_key = os.environ.get("GROQ_API_KEY")

model = ChatGroq(
    model="meta-llama/llama-4-scout-17b-16e-instruct",
    api_key=api_key
)

# Setup embedding & FAISS
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

def get_or_create_faiss(path: str, embedding):
    if os.path.exists(path):
        return FAISS.load_local(path, embedding, allow_dangerous_deserialization=True)
    else:
        db = FAISS.from_texts(["placeholder"], embedding)
        db.save_local(path)
        return db

# Inisialisasi FAISS
db_path = "vectorstore/faiss_index"
db = get_or_create_faiss(db_path, embedding)

# === Awal Percakapan ===
messages = [
    SystemMessage(content=(
        "Kamu adalah tutor belajar pribadi untuk siswa SMK, khususnya di bidang Akuntansi. "
        "Berikan penjelasan yang mudah dipahami, terstruktur, dan relevan."
    ))
]

# === Loop Chat ===
print("üéì AI Tutor Akuntansi SMK ‚Äî ketik 'exit' untuk berhenti")
while True:
    user_input = input("üë®‚Äçüéì Kamu: ")

    if user_input.strip().lower() == "exit":
        print("üëã Sampai jumpa!")
        break

    # Ambil konteks dari FAISS
    docs = db.similarity_search(user_input, k=2)
    if docs:
        retrieved_content = "\n\n".join([doc.page_content for doc in docs])
        messages.append(SystemMessage(content=f"Berikut adalah informasi yang pernah kamu pelajari sebelumnya:\n\n{retrieved_content}"))

    # Tambahkan pesan user
    messages.append(HumanMessage(content=user_input))

    # Dapatkan respon AI
    res = model.invoke(messages)
    display(Markdown("ü§ñ **Tutor:**\n\n" + res.content))

    # Tambahkan respon AI ke history dan FAISS jika ada perintah "ingat"/"catat"
    messages.append(AIMessage(content=res.content))
    if "catat" in user_input.lower() or "ingat" in user_input.lower():
        doc = Document(page_content=res.content)
        db.add_documents([doc])
        db.save_local(db_path)
        print("üß† Disimpan ke memori FAISS.")
```

```python
from dotenv import load_dotenv
from langchain_groq import ChatGroq
from langchain_core.messages import SystemMessage, HumanMessage, AIMessage
from langchain.vectorstores import FAISS
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.schema import Document
from pathlib import Path
import os
import sys
import time

# Untuk menampilkan di Jupyter
try:
    from IPython.display import display, Markdown, clear_output
    IN_JUPYTER = True
except ImportError:
    IN_JUPYTER = False

# Untuk animasi loading
from threading import Thread, Event

# ========== Setup ==========
load_dotenv()
api_key = os.environ.get("GROQ_API_KEY")

model = ChatGroq(
    model="meta-llama/llama-4-scout-17b-16e-instruct",
    api_key=api_key
)

embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# ========== FAISS ==========
def get_or_create_faiss(path: str, embedding):
    if os.path.exists(path):
        return FAISS.load_local(path, embedding, allow_dangerous_deserialization=True)
    else:
        db = FAISS.from_texts(["placeholder"], embedding)
        db.save_local(path)
        return db

db_path = "vectorstore/faiss_index"
db = get_or_create_faiss(db_path, embedding)

# ========== Loading Spinner ==========
class Spinner:
    def __init__(self, message="‚è≥ Thinking..."):
        self.message = message
        self.stop_running = Event()

    def spin(self):
        while not self.stop_running.is_set():
            for cursor in "|/-\\":
                print(f"\r{self.message} {cursor}", end="", flush=True)
                time.sleep(0.1)

    def __enter__(self):
        self.thread = Thread(target=self.spin)
        self.thread.start()

    def __exit__(self, exc_type, exc_value, traceback):
        self.stop_running.set()
        self.thread.join()
        print("\r", end="")

# ========== Riwayat Awal ==========
messages = [
    SystemMessage(content=(
        "Kamu adalah tutor belajar pribadi untuk siswa SMK, khususnya di bidang Akuntansi. "
        "Berikan penjelasan yang mudah dipahami, terstruktur, dan relevan."
    ))
]

# ========== Mulai Chat ==========
print("üéì AI Tutor Akuntansi SMK ‚Äî ketik 'exit' untuk berhenti")
while True:
    user_input = input("üë®‚Äçüéì Kamu: ").strip()

    if user_input.lower() == "exit":
        print("üëã Sampai jumpa!")
        break

    # Cari konteks dari FAISS
    docs = db.similarity_search(user_input, k=2)
    if docs:
        retrieved = "\n\n".join([doc.page_content for doc in docs])
        messages.append(SystemMessage(content=f"Berikut adalah informasi yang pernah kamu pelajari sebelumnya:\n\n{retrieved}"))

    messages.append(HumanMessage(content=user_input))

    # Tampilkan animasi loading
    with Spinner():
        stream = model.stream(messages)

        collected = []
        for chunk in stream:
            if chunk.content:
                collected.append(chunk.content)
                if IN_JUPYTER:
                    clear_output(wait=True)
                    display(Markdown("ü§ñ **Tutor:**\n\n" + "".join(collected)))
                else:
                    print("".join(collected), end="\r", flush=True)

    final_response = "".join(collected)
    if not IN_JUPYTER:
        print("\nü§ñ Tutor:\n", final_response)

    messages.append(AIMessage(content=final_response))

    # Simpan ke FAISS jika diminta
    if any(trigger in user_input.lower() for trigger in ["catat", "ingat"]):
        db.add_documents([Document(page_content=final_response)])
        db.save_local(db_path)
        print("üß† Disimpan ke memori FAISS.")
```
